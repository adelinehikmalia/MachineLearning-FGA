# Train dan Test Generator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_data = ImageDataGenerator(
    rescale = 1/255,
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    shear_range = 0.2,
    fill_mode = 'nearest',
    validation_split = 0.2)
```

```python
batch_size = 256
train_data_gen = train_data.flow_from_directory(
    data,
    target_size = (150, 150),
    batch_size = batch_size,
    class_mode = 'categorical',
    subset = 'training')

val_data_gen = train_data.flow_from_directory(
    data,
    target_size = (150, 150),
    batch_size = batch_size,
    class_mode = 'categorical',
    subset = 'validation')
```

# Membangun Model

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3, 3), activation = 'relu', input_shape = (150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(512, activation = 'relu'),
    tf.keras.layers.Dense(21, activation = 'softmax')
])
model.summary()
```

```python
tf.keras.utils.plot_model(
    model,
    show_shapes = True,
    show_layer_names = True,
)
```

```python
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']               
)
```

# Callbacks

```python
# misal akurasi threshold 92%
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs = None):
        if logs.get('accuracy')>0.92:
            print("\nAkurasi telah mencapai > 92%!")
            self.model.stop_training = True
```

```python
print(train_data_gen.samples // batch_size)
print(val_data_gen.samples // batch_size)
```

```python
history = model.fit(
    train_data_gen,
    epochs = 20,
    steps_per_epoch = 31,
    validation_data = val_data_gen,
    validation_steps = 7,
    verbose = 1,
    callbacks = [myCallback()])
```

# Evaluasi Model

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Akurasi')
plt.ylabel('Akurasi')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
  
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
```

# Konversi Model

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
```
